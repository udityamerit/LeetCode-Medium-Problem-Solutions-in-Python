```python
from collections import defaultdict

class Solution:
    def countCompleteSubarrays(self, nums: List[int]) -> int:
        total_unique = len(set(nums))
        count = 0
        n = len(nums)
        
        for i in range(n):
            unique = set()
            for j in range(i, n):
                unique.add(nums[j])
                # print(unique)
                if len(unique) == total_unique:
                    count += 1
        return count
```
