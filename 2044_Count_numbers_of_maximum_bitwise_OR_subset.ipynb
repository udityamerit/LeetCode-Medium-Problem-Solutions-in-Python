```python
class Solution:
    def countMaxOrSubsets(self, nums: List[int]) -> int:
        n = len(nums)
        mx = []
    
        for r in range(1, n + 1):  
            for subset in combinations(nums, r):
                subset_or = 0
                for num in subset:
                    subset_or |= num  
                mx.append(subset_or)
                print(subset)
        return mx.count(max(mx))
                
```
