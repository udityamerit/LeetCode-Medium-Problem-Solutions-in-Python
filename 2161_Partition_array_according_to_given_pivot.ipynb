#### Solution 1 in Python

```python
class Solution:
    def pivotArray(self, nums: List[int], pivot: int) -> List[int]:
        l = [i for i in nums if i<pivot]
        print(l)
        m = [i for i in nums if i> pivot]
        print(m)
        n = [i for i in nums if i==pivot]
        print(n)
        return l+n+m
```
