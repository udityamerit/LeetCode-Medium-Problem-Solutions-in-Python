#### Solution 1 

```python
class Solution:
    def maximumUniqueSubarray(self, nums: List[int]) -> int:
      # solution 1 
        l = []
        ans = 0
        d = len(set(nums))
        # print(d)
        for i in range(len(nums)):
            c = []
            start = i
            for j in range(i,len(nums)):
                end = j
                for k in range(start, end+1):
                    c.append(nums[k])
                # print(c)
                if len(c)==d:
                    if set(c) not in l:
                        l.append(set(c))
                    if ans<sum(set(c)):
                        ans = sum(set(c))

                c = []
        # print(ans)
        print(l)
        return ans
```

#### Solution 2

```python
class Solution:
    def maximumUniqueSubarray(self, nums: List[int]) -> int:     
        # Solution 2

        seen = set()
        left = 0
        curr_sum = 0
        max_sum = 0
        
        for right in range(len(nums)):
            while nums[right] in seen:
                seen.remove(nums[left])
                curr_sum -= nums[left]
                left += 1
                
            seen.add(nums[right])
            curr_sum += nums[right]
            max_sum = max(max_sum, curr_sum)
        return max_sum

```
