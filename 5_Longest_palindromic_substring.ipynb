#### Solution 1 in python

```python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        l = ""   
        n = len(s)
        if n==1:
            return s
        for k in range(1, n + 1):
            for i in range(n - k + 1):
                poll = s[i:i+k]
                if poll[::]==poll[::-1]:
                    if len(l)<len(poll):
                        l = poll
        print(l)
        return l
```
