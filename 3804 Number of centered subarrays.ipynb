#### Solution 1

```python
class Solution:
    def centeredSubarrays(self, nums: List[int]) -> int:
        subarrays = []
        for i in range(len(nums)):        
            for j in range(i + 1, len(nums) + 1): 
                subarray = nums[i:j]  
                if sum(subarray) in subarray:    
                    subarrays.append(subarray)
        return len(subarrays)
    
    
```
