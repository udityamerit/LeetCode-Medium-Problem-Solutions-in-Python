#### Solution 1 in python

```python
class Solution:
    def countSubstrings(self, s: str) -> int:
        l = 0
        n = len(s)
        for k in range(1,n+1):
            for i in range(n-k+1):
                poll = s[i:i+k]
                if poll[::]==poll[::-1]:
                    l = l+1
        return l
        
```
