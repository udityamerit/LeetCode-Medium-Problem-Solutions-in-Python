```python
class Solution:
    def countSubarrays(self, nums: List[int], k: int) -> int:
        ## the time complexity of this code is n3, which is not efficient for large inputs.
        # n = len(nums)
        # m = max(nums)
        # a = 0
        # for b in range(k, n+1):
        #     for i in range(n-b+1):
        #         subarr = nums[i:i+b]
        #         # print(subarr)
        #         if subarr.count(m)>=k:
        #             a = a+1
        # return a

        max_num = max(nums)
        n = len(nums)
        count = 0
        left = 0
        max_count = 0
        
        for right in range(n):
            if nums[right] == max_num:
                max_count += 1

            while max_count >= k:
                count += (n - right)  
                if nums[left] == max_num:
                    max_count -= 1
                left += 1

        return count
```
