```python
class Solution:
    def maxRotateFunction(self, nums: List[int]) -> int:
        # question no 396

        # nums = [4,3,2,6]
        # n = []
        # ans = -inf
        # def rev(arr, n):
        #         arr.insert(0, arr[len(arr)-1])
        #         arr.pop()
        #         n = arr
        #         return n
                

        # for i in range(len(nums)):
        #         c = rev(nums, n)
        #         d = 0
        #         for i in range(len(c)):
        #                 d = d+(c[i]*i)
        #                 # print(d)
        #         if ans<d:
        #                 ans = d
        #                 # print(ans)
        # return ans
        
        n = len(nums)
        total_sum = sum(nums)
        F = sum(i * num for i, num in enumerate(nums))
        
        max_value = F
        for k in range(1, n):
            F = F + total_sum - n * nums[n - k]
            max_value = max(max_value, F)
        
        return max_value

            
```
